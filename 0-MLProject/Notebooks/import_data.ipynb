# ***SPOTIFY DATA***
### Our columns here:
- name - Name of song
- album - Name of album
- artist - Name of artist/s involved
- release_date - Release date of album
- length - Song length in milliseconds
- popularity - Percent popularity of the song based on Spotify's algorithm (possibly the number of stream at a certain period of time)
- danceability - How suitable a track is for dancing based on a combination of musical elements including tempo, rhythm stability, beat strength, and overall regularity
- acousticness - How acoustic a song is
- energy - A perceptual measure of intensity and activity
- instrumentalness - Te amount of vocals in the song
- liveness - Probability that the song was recorded with a live audience
- loudness - Tdency o of music to be recorded at steadily higher volumes
- speechiness - Presence of spoken words in a track (if the speechiness of a song is above 0.66, it is probably made of spoken words, a score between 0.33 and 0.66 is a song that may contain both music and words, and a score below 0.33 means the song does not have any speech)
- valence - A measure of how happy or sad the song sounds
- tempo - Beats per minute

```python
# Histograms for YouTube metrics
plt.figure(figsize=(15, 10))
#kde para suavizar la lectura
# viewCount
plt.subplot(2, 2, 1)
sns.histplot(youtailor['viewCount'], bins=30, kde=True, color='lightgreen')
plt.title('Distribution of View Count on YouTube')

# likeCount
plt.subplot(2, 2, 2)
sns.histplot(youtailor['likeCount'], bins=30, kde=True, color='lightgreen')
plt.title('Distribution of Like Count on YouTube')

# dislikeCount
plt.subplot(2, 2, 3)
sns.histplot(youtailor['dislikeCount'], bins=10, kde=True, color='lightgreen')
plt.title('Distribution of Dislike Count on YouTube')

# commentCount
plt.subplot(2, 2, 4)
sns.histplot(youtailor['commentCount'], bins=30, kde=True, color='lightgreen')
plt.title('Distribution of Comment Count on YouTube')

plt.tight_layout()
plt.show()
```

```python
# Correlation Matrix
plt.figure(figsize=(10, 8))
correlation_matrix = youtailor[['viewCount', 'likeCount', 'dislikeCount', 'commentCount']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Between YouTube Metrics')
plt.show()
```

```python
# Scatter plot for Likes vs Dislikes

plt.figure(figsize=(10, 8))
sns.scatterplot(x='likeCount', y='commentCount', data=youtailor, color='blue')
plt.title('Relationship Between Likes and Comments on YouTube')
plt.xlabel('Like Count')
plt.ylabel('Comment Count')
plt.show()
```
